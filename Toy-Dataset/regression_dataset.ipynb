```python
# now lets train our model with different ml models
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
lsso = Lasso()
ridgeregression  = Ridge()
lr = LinearRegression()
```

```python
ourdf.columns
ourdf.drop(columns=["Age_Group","Skill_Score_Normalized"],inplace=True)
ourdf
```

```python
# lets do train test split on our dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(ourdf.iloc[:,:-1],ourdf.iloc[:,-1],test_size=0.8,random_state=1)
```

```python
# X_test,X_train
```

```python
# now lets train our differet model 
lr.fit(X_train,y_train)
```

```python
lsso.fit(X_train,y_train)
```

```python
ridgeregression.fit(X_train,y_train)
```

```python
# lets show the intercept and cofeicent of our different model 
print(f"cofecient for lasso regression: \n{lsso.coef_}")
print(f"cofecient for Ridge regression: \n{ridgeregression.coef_}")
print(f"cofecient for Linear regression: \n{lr.coef_}")
```

```python
# intercept values are 
print(f"intercept for lasso regression: \n{lsso.intercept_}")
print(f"intercept for Ridge regression: \n{ridgeregression.intercept_}")
print(f"intercept for Linear regression: \n{lr.intercept_}")
```

```python
# letrs compare the r2 square for all our models
from sklearn.metrics import r2_score
print(f"r2 score for linear regression {r2_score(y_test,lr.predict(X_test))}")
print(f"r2 score for lasso {r2_score(y_test,lsso.predict(X_test))}")
print(f"r2 score for linear regression {r2_score(y_test,lr.predict(X_test))}")
```

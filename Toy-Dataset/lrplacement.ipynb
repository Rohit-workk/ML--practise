```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv("placement.csv")
df
```

```python
plt.scatter(df.cgpa,df.package,color = "red")
plt.xlabel("independent variable\ncgpa")
plt.ylabel("dependent variable\npackage")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_trian,X_test,y_train,y_test = train_test_split(df[["cgpa"]],df.package, test_size= 0.8,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_trian,y_train)
```

```python
lr.predict(X_test)
```

```python
plt.scatter(df.cgpa,df.package,color = "red")
plt.plot(X_test,lr.predict(X_test),color = "black")
plt.xlabel("independent variable\ncgpa")
plt.ylabel("dependent variable\npackage")
plt.show()
```

```python
# calculate mse,rootmeansqure,mean absolute ,r2_score
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"mean square error {(np.sum((y_test -lr.predict(X_test))*2))/len(y_test)}")
print(f"mean square error {mean_squared_error(y_test,lr.predict(X_test))}")
print()
```

```python
print(f"root mean square error {np.sqrt((np.sum((y_test -lr.predict(X_test))*2))/len(y_test))}")
print(f"mean square error {np.sqrt(mean_squared_error(y_test,lr.predict(X_test)))}")
print()
```

```python
# mean absolute error 
print(f"mean absolute error {mean_absolute_error(y_test,lr.predict(X_test))}")
```

```python
# r2 score 
print(f"r2 score{r2_score(y_test , lr.predict(X_test))}")
```

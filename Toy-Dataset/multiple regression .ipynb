```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples=100,n_features=5,n_targets=1,noise=20)
```

```python
# make it a dataframe 
df = pd.DataFrame({
    "col1":x[:,0],
    "col2":x[:,1], 
    "col3":x[:,3],
    "col4":x[:,4], 
    "col5":x[:,2],
    "target":y
})
```

```python
df
```

```python
x.shape
```

```python
# lets split our datatset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns="target"),df['target'],test_size=0.2,random_state=2)
x_train
```

```python
# now lets train our model 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print("coef  ",lr.intercept_)
print("intercept means b1,b2,b3,b4 ",lr.coef_)
```

```python
# now predict values 
pred_val = lr.predict(x_test)
```

```python
# metrices for our multiple linear regression model 
from sklearn.metrics import mean_absolute_error,r2_score
print("mean absolute eror ",mean_absolute_error(y_test,pred_val))
print("r2 score ",r2_score(y_test,pred_val))
```

## ridge regression 

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=5)

```

```python
rr.fit(x_train,y_train)
```

```python
# intercept and coef with rr
print("slope is ",rr.intercept_)
print("b1,b2,b3,b4,b5",rr.coef_)
```

```python
# with linear model 
print("coef  ",lr.intercept_)
print("intercept means b1,b2,b3,b4 ",lr.coef_)
```

# now metrices comparsion 

```python
print("r2 score for lr ",r2_score(y_test,lr.predict(x_test)))
print("r2score for rr",r2_score(y_test,rr.predict(x_test)))
```

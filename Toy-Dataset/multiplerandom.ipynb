```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_features=2,n_targets=1,noise=20,n_samples=100)
```

```python
# making of object
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# training of model 
lr.fit(x_train,y_train)

```

```python
pred_y = lr.predict(x_test)
```

```python
class gd:
    def __init__(self, learningrate=0.00001, epoch=100):
        self.ep = epoch
        self.lr = learningrate
        self.coef = None
        self.intercept = None

    def fit(self, x_train, y_train):
        # initialize intercept (B0)
        self.intercept = 0
        # initialize coef
        self.coef = np.ones(x_train.shape[1])

        # loop for number of epochs
        for i in range(self.ep):
            # prediction
            y_cap = np.dot(x_train, self.coef) + self.intercept

            # gradient for intercept
            intercept_derivative = -2 * np.mean(y_train - y_cap)
            self.intercept = self.intercept + self.lr * intercept_derivative

            # gradient for coef
            coef_derivative = (-2 * np.dot(x_train.T, (y_train - y_cap))) / x_train.shape[0]
            self.coef = self.coef + (self.lr * coef_derivative)

        return (self.intercept, self.coef)

    def predict(self, x_test):
        return np.dot(x_test, self.coef) + self.intercept



```

```python
# training
ourmodel = gd()
ourmodel.fit(x_train, y_train)
```

# classfication alsgorithm practise

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# lets make a dataset or import classsfication dataset
from sklearn.datasets import make_classification
X, y = make_classification(
    n_classes=2,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_repeated=0,
    n_clusters_per_class=1,
    random_state=42,
    class_sep=0.5
)
```

```python
# now lets plot this datasetabs
# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of Classification Dataset")
plt.show()
```

```python
#  makining model object 
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
```

```python
# train test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
y
```

```python
# now fit our model
logr.fit(x_train,y_train)
```

```python
logr.coef_
```

```python
logr.intercept_
```

```python
# now lets make the line
x_points = np.linspace(-2,2,100)
```

```python
m =-( logr.coef_[0][0]/logr.coef_[0][1])
b = -(logr.intercept_[0]/logr.coef_[0][1])
```

```python
y_points= m*x_points+b
```

```python
# Scatter plot
plt.plot(x_points,y_points)
plt.ylim(-3,3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of Classification Dataset")
plt.show()
```

```python
# now classfication metrics 
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score
print(f"accuracy score {accuracy_score(y_test,logr.predict(x_test))}")
print(f"confusion matrix \n{confusion_matrix(y_test,logr.predict(x_test))}")
print(f"precision score {precision_score(y_test,logr.predict(x_test))}")
print(f"f1score {f1_score(y_test,logr.predict(x_test))}")
 
 
```
